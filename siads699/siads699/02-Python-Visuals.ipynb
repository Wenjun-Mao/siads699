```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('assets/log_df.csv')
```

```python
# count the number of status records for verification
df['status'].value_counts()
```

```python
plt.title('Accuracy of Model')
plt.xlabel('Status')
plt.ylabel('Number of Queries')
# create a bar chart of the status counts
df['status'].value_counts().plot(kind='bar')
df.rename(columns={'status':'Status'}, inplace=True)
plt.xticks([0,1,2],['No Error','Irrelevant Question','Wrong Code Generated'])
plt.show()
```
